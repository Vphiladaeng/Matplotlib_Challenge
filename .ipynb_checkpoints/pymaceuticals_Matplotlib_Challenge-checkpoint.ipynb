```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_data = pd.merge(clinical_data, mouse_data, how = "left", on = "Mouse ID")

# Display the data table for preview
merge_data
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor_volume = merge_data.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
avg_tumor_df = pd.DataFrame(avg_tumor_volume)
avg_tumor_df = avg_tumor_df.reset_index()

# Preview DataFrame
avg_tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error_tumor_volume = merge_data.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
sem_tumor_df = pd.DataFrame(standard_error_tumor_volume)
sem_tumor_df = sem_tumor_df.reset_index()

# Preview DataFrame
sem_tumor_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_pivottable = avg_tumor_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
avg_pivottable
```

```python
# Drop the drugs those are not needed based on the treatment plot
cleaned_tumorv_mean_pivot = avg_pivottable.drop(['Ceftamin','Naftisol','Propriva','Ramicane','Stelasyn','Zoniferol'], 
                                                         axis=1)
cleaned_tumorv_mean_pivot
```

```python
cleaned_tumorv_se_pivot = avg_pivottable.drop(['Ceftamin','Naftisol','Propriva','Ramicane','Stelasyn','Zoniferol'], 
                                                         axis=1)
cleaned_tumorv_se_pivot
```

```python
# Set up the x_axis,y_axis data and standard error for the plot
x_axis = cleaned_tumorv_mean_pivot.index.values

# Setting up the plot for Capomulin
se = cleaned_tumorv_se_pivot["Capomulin"]
capo = plt.errorbar(x_axis, cleaned_tumorv_mean_pivot["Capomulin"], color='red',yerr=se, fmt="o", ls="dashed", linewidth=0.5, alpha=1)

# Setting up the plot for Infubinol
se = cleaned_tumorv_se_pivot["Infubinol"]
infu = plt.errorbar(x_axis, cleaned_tumorv_mean_pivot["Infubinol"],color='blue', yerr=se, fmt="^", ls="dashed",linewidth=0.5, alpha=1)

# Setting up the plot for Ketapril
se = cleaned_tumorv_se_pivot["Ketapril"]
keta = plt.errorbar(x_axis, cleaned_tumorv_mean_pivot["Ketapril"],color='green', yerr=se, fmt="s", ls="dashed",linewidth=0.5,  alpha=1)

# Setting up the plot for Ketapril
se = cleaned_tumorv_se_pivot["Placebo"]
place = plt.errorbar(x_axis, cleaned_tumorv_mean_pivot["Placebo"],color='black', yerr=se, fmt="d", ls="dashed",linewidth=0.5,  alpha=1)

# Set up the x_axis and y_axis limits
plt.ylim(33,73)
plt.xlim(-2, 48)

# Set up the title,x_axis and y_axis labels
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Set up the grid for y axis
plt.grid(axis='y')

# Set up the legend
plt.legend((capo, infu, keta, place), ("Capomulin", "Infubinol", "Ketapril", "Placebo"),loc="upper left")

# Save the Figure
plt.savefig("TumorTreatment.png")

# Display the plot
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
ms_mean = merge_data.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
ms_mean

# Convert to Dataframe
ms_mean_df = pd.DataFrame(ms_mean).reset_index()

# Preview Dataframe
ms_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
ms_se = merge_data.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
ms_se

# Convert to DataFrame
ms_se_df = pd.DataFrame(ms_se).reset_index()

# Preview DataFrame
ms_se_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
ms_mean_df_pivot = ms_se_df.pivot(index ="Timepoint", values = "Metastatic Sites", columns = "Drug")
ms_se_df_pivot = ms_se_df.pivot(index ="Timepoint", values = "Metastatic Sites", columns = "Drug")

# Preview that Reformatting worked
ms_mean_df_pivot.head()
#ms_se_df_pivot.head()
```

```python
# Set up the x_axis,y_axis data and standard error for the plot
x_axis = ms_mean_df_pivot.index.values

# Setting up the plot for Capomulin
se = ms_mean_df_pivot["Capomulin"]
capo = plt.errorbar(x_axis, ms_mean_df_pivot["Capomulin"], color='red',yerr=se, fmt="o", ls="dashed", linewidth=0.5, alpha=1)

# Setting up the plot for Infubinol
se = ms_mean_df_pivot["Infubinol"]
infu = plt.errorbar(x_axis, ms_mean_df_pivot["Infubinol"], color='blue', yerr=se, fmt="^", ls="dashed",linewidth=0.5, alpha=1)

# Setting up the plot for Ketapril
se = ms_mean_df_pivot["Ketapril"]
keta = plt.errorbar(x_axis, ms_mean_df_pivot["Ketapril"], color='green', yerr=se, fmt="s", ls="dashed",linewidth=0.5,  alpha=1)

# Setting up the plot for Ketapril
se = ms_mean_df_pivot["Placebo"]
place = plt.errorbar(x_axis, ms_mean_df_pivot["Placebo"], color='black', yerr=se, fmt="d", ls="dashed",linewidth=0.5,  alpha=1)

# Set up the title,x_axis and y_axis labels
plt.title("Metastatic Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Drug")

# Set up the grid for y axis
plt.grid(axis = 'y')

# Set up the legend
plt.legend((capo, infu, keta, place), ("Capomulin", "Infubinol", "Ketapril", "Placebo"),loc="upper left")

# Save the Figure
plt.savefig("MetastaticTreatment.png")

# Display the plot
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = merge_data.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]
mice_count

# Convert to Dataframe
mice_count_df = pd.DataFrame(mice_count)
mice_count_df = mice_count_df.reset_index()
mice_count_df = mice_count_df.rename(columns = {"Mouse ID":"Mouse Count"})

# Preview Dataframe
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Dataframes
mice_count_df_pivot = mice_count_df.pivot(index ="Timepoint", values = "Mouse Count", columns = "Drug")

# Preview the Dataframe
mice_count_df_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)

# Set up the x_axis,y_axis data and standard error for the plot
x_axis = mice_count_df_pivot.index.values

# Setting up the plot for Capomulin
se = mice_count_df_pivot["Capomulin"]
capo = plt.errorbar(x_axis, mice_count_df_pivot["Capomulin"], color='red',yerr=se, fmt="o", ls="dashed", linewidth=0.5, alpha=1)

# Setting up the plot for Infubinol
se = mice_count_df_pivot["Infubinol"]
infu = plt.errorbar(x_axis, mice_count_df_pivot["Infubinol"], color='blue', yerr=se, fmt="^", ls="dashed",linewidth=0.5, alpha=1)

# Setting up the plot for Ketapril
se = mice_count_df_pivot["Ketapril"]
keta = plt.errorbar(x_axis, mice_count_df_pivot["Ketapril"], color='green', yerr=se, fmt="s", ls="dashed",linewidth=0.5,  alpha=1)

# Setting up the plot for Ketapril
se = mice_count_df_pivot["Placebo"]
place = plt.errorbar(x_axis, mice_count_df_pivot["Placebo"], color='black', yerr=se, fmt="d", ls="dashed",linewidth=0.5,  alpha=1)

# Set up the title,x_axis and y_axis labels
plt.title("Metastatic Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Drug")

# Set up the grid for y axis
plt.grid(axis = 'y')

# Set up the legend
plt.legend((capo, infu, keta, place), ("Capomulin", "Infubinol", "Ketapril", "Placebo"),loc="upper left")

# Save the Figure
plt.savefig("MetastaticSpreadDuringTreatment.png")

# Display the plot
plt.show()
```
